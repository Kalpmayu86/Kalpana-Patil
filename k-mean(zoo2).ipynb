```python
# Normalization function 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled__df = scaler.fit_transform(df.iloc[:,1:])
```

```python
from sklearn.datasets import make_blobs
import pandas as pd

dataset, classes = make_blobs(n_samples=100, n_features=17, centers=4, cluster_std=0.5, random_state=0)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(scaled__df)
```

```python
kmeans.labels_  # same as kmeans.predict(df)
```

```python
kmeans.inertia_
```

```python
kmeans.cluster_centers_
```

```python
import matplotlib.pyplot as plt
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled__df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(KMeans(),k=6)
visualizer.fit(scaled__df)
visualizer.show()
```

# using pca

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=4)
pca_df=pca.fit_transform(scaled__df)
pca_df_new = pd.DataFrame(data=pca_df, columns = ['components1','components2','components3','components4'])
pca_df_new
```

```python
pca_df_new['clusterid_new'] = clusters_new.labels_
```

```python
from yellowbrick.cluster import SilhouetteVisualizer
 
fig, ax = plt.subplots(2, 2, figsize=(15,8))
for i in [2, 3, 4, 5]:
    '''
    Create KMeans instance for different number of clusters
    '''
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=42)
    q, mod = divmod(i, 2)
    '''
    Create SilhouetteVisualizer instance with KMeans instance
    Fit the visualizer
    '''
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(pca_df_new) 
```
